```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset with a different encoding
data = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='latin1')
```

```python
# Display basic information about the dataset
print(data.info())
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Summary statistics
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Explore the 'country_txt' column to identify the most affected countries
country_counts = data['country_txt'].value_counts()
top_countries = country_counts.head(10)
print(top_countries)
```

```python
# Visualize the top affected countries
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.title('Top 10 Countries with Highest Terrorism Incidents')
plt.xlabel('Country')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```

```python
# Explore 'region_txt' column to identify the most affected regions
region_counts = data['region_txt'].value_counts()
top_regions = region_counts.head(5)
print(top_regions)
```

```python
# Visualize the top affected regions
plt.figure(figsize=(10, 6))
sns.barplot(x=top_regions.index, y=top_regions.values, palette='magma')
plt.title('Top 5 Regions with Highest Terrorism Incidents')
plt.xlabel('Region')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()

```

```python
# Explore 'city' column to identify the most affected cities
city_counts = data['city'].value_counts()
top_cities = city_counts.head(10)
print(top_cities)
```

```python
# Visualize the top affected cities
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='plasma')
plt.title('Top 10 Cities with Highest Terrorism Incidents')
plt.xlabel('City')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```
